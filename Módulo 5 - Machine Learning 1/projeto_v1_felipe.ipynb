#### Tratamendo das variáveis categóricas

```python
from sklearn.preprocessing import OneHotEncoder

cols_to_transform = ['Department', 'EducationField', 'JobRole', 'MaritalStatus']

onehotencoder = OneHotEncoder(sparse=False)

transformed = onehotencoder.fit_transform(df[cols_to_transform])
df_transformed = pd.DataFrame(transformed, columns=onehotencoder.get_feature_names(cols_to_transform))

df = pd.concat([df, df_transformed], axis=1)

df.drop(cols_to_transform, axis=1, inplace=True)
```

```python
df.head(2)
```

```python
df.BusinessTravel.value_counts()
```

```python
df['Gender'] = df['Gender'].replace({'Male': 0, 'Female': 1})
df['BusinessTravel'] = df['BusinessTravel'].replace({'Non-Travel': 0, 'Travel_Rarely': 1, 'Travel_Frequently':2})
df['OverTime'] = df['OverTime'].replace({'Yes':1, 'No':0})
```

```python
df.head(2)
```

```python
corr_to_analysis = []

corr = df.corr(method='pearson')

for item in corr.columns:
    strong_corr = zip(corr[item].abs().sort_values(ascending=False)[1:4].index, corr[item].abs().sort_values(ascending=False)[1:4].values)
    for index, value in strong_corr:
        if value >= 0.75 and ((item, index, value) not in corr_to_analysis and (index, item, value) not in corr_to_analysis):
            corr_to_analysis.append((item, index, value))
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# seleciona as colunas do DataFrame
cols = ['OverTime', 'JobLevel', 'MonthlyIncome', 'YearsAtCompany', 'Age', 'Attrition']

# cria um PairGrid com as colunas selecionadas
g = sns.PairGrid(df[cols], hue='Attrition')

# define o gráfico para a diagonal
g.map_diag(plt.hist)

# define o gráfico para a parte inferior da diagonal
g.map_lower(sns.kdeplot, levels=2)

# define o gráfico para a parte superior da diagonal
g.map_upper(plt.scatter)

# adiciona a legenda
g.add_legend()

# exibe o plot
plt.show()
```

- 'Overtime' - Praticamente metade dos que fizeram hora-extra se demitiu; Já entre os que não fizeram, poucos se demitiram;
- 'JobLevel' - Em geral, quanto menor o nível de senioridade, maior a chance de se demitir;
- 'MonthlyIncome' - Quem recebe menos, sai mais facilmente;
- 'YearsAtCompany' - Quanto menos tempo de empresa, maior a chance de sair;
- 'Age' - Não é tão nítido, mas quanto menor a idade, maior a chance de sair;
